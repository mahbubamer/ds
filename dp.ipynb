```python
# =========================
# 1) DP - Fibonacci Sequence
# =========================

def fib_dp(n):
    # Returns nth Fibonacci number using DP (bottom-up)
    # fib(0)=0, fib(1)=1
    if n <= 1:
        return n

    dp = [0] * (n + 1)      # dp[i] will store fib(i)
    dp[0] = 0
    dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]


# =========================
# 2) DP - LCS (Longest Common Subsequence)
# =========================

def lcs_dp(X, Y):
    # Returns (length, one LCS string) using DP
    m = len(X)
    n = len(Y)

    # dp[i][j] = LCS length of X[0..i-1] and Y[0..j-1]
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill dp table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if X[i - 1] == Y[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # Reconstruct one LCS by backtracking
    i, j = m, n
    lcs_chars = []
    while i > 0 and j > 0:
        if X[i - 1] == Y[j - 1]:
            lcs_chars.append(X[i - 1])
            i -= 1
            j -= 1
        else:
            # Move to the direction of the bigger value
            if dp[i - 1][j] >= dp[i][j - 1]:
                i -= 1
            else:
                j -= 1

    lcs_string = "".join(reversed(lcs_chars))
    return dp[m][n], lcs_string


# ==========================================
# 3) DP - Coin Change (Total Number of Ways)
# ==========================================

def coin_change_ways(coins, amount):
    # Count number of ways to make 'amount' using unlimited coins
    # Order does NOT matter (classic DP)
    dp = [0] * (amount + 1)
    dp[0] = 1  # 1 way to make 0: choose nothing

    for coin in coins:
        for a in range(coin, amount + 1):
            dp[a] += dp[a - coin]

    return dp[amount]


# =========================================
# 4) DP - Coin Change (Minimum Number of Coins)
# =========================================

def coin_change_min_coins(coins, amount):
    # Minimum coins to make 'amount' using unlimited coins
    # Returns -1 if not possible
    INF = 10**9
    dp = [INF] * (amount + 1)
    dp[0] = 0

    for a in range(1, amount + 1):
        for coin in coins:
            if a - coin >= 0:
                dp[a] = min(dp[a], 1 + dp[a - coin])

    return -1 if dp[amount] == INF else dp[amount]


# =========================
# Example Runs (You can change these)
# =========================
if __name__ == "__main__":
    # Fibonacci
    n = 10
    print("Fibonacci DP:", n, "->", fib_dp(n))

    # LCS
    X = "AGGTAB"
    Y = "GXTXAYB"
    length, seq = lcs_dp(X, Y)
    print("LCS length:", length)
    print("LCS sequence:", seq)

    # Coin Change
    coins = [1, 2, 5]
    amount = 5
    print("Coin Change Ways:", coin_change_ways(coins, amount))
    print("Coin Change Min Coins:", coin_change_min_coins(coins, amount))
```

```python
def fib_dp(n):                 # function to find Fibonacci of n using DP
    if n <= 1:                 # base cases: fib(0)=0, fib(1)=1
        return n               # return directly for 0 or 1

    dp = [0] * (n + 1)         # dp array size n+1, dp[i] will store fib(i)
    dp[0] = 0                  # store base value fib(0)
    dp[1] = 1                  # store base value fib(1)

    for i in range(2, n + 1):  # compute fib(2) up to fib(n)
        dp[i] = dp[i-1] + dp[i-2]  # DP relation: fib(i)=fib(i-1)+fib(i-2)

    return dp[n]               # final answer stored at dp[n]


n = 10                         # example input
print(fib_dp(n))               # prints fib(10)
```

```python
def lcs_dp(X, Y):                                  # function to find LCS length of X and Y
    m = len(X)                                     # length of first string
    n = len(Y)                                     # length of second string

    dp = [[0] * (n + 1) for _ in range(m + 1)]     # dp table (m+1) x (n+1), filled with 0

    for i in range(1, m + 1):                      # i represents first i chars of X
        for j in range(1, n + 1):                  # j represents first j chars of Y

            if X[i - 1] == Y[j - 1]:               # if current characters match
                dp[i][j] = 1 + dp[i - 1][j - 1]    # take diagonal + 1 (extend subsequence)

            else:                                   # if characters do NOT match
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])  
                # choose best of:
                # 1) removing one char from X (up)
                # 2) removing one char from Y (left)

    return dp[m][n]                                 # final LCS length is in last cell


X = "AGGTAB"                                       # example string 1
Y = "GXTXAYB"                                      # example string 2
print(lcs_dp(X, Y))                                # prints LCS length
```

```python
def coin_change_ways(coins, amount):    # function to count ways to form 'amount'
    dp = [0] * (amount + 1)             # dp[a] stores number of ways to make amount a
    dp[0] = 1                           # 1 way to make 0 (choose no coin)

    for coin in coins:                  # take each coin one by one
        for a in range(coin, amount + 1):   # update dp from coin value up to amount
            dp[a] += dp[a - coin]       # add ways to make (a-coin), then use this coin

    return dp[amount]                   # answer: number of ways to make 'amount'


coins = [1, 2, 5]                       # coin denominations
amount = 5                              # target
print(coin_change_ways(coins, amount))  # prints total ways
```

```python
def coin_change_min(coins, amount):     # function to find minimum coins for 'amount'
    INF = 10**9                         # a very large number (acts like infinity)
    dp = [INF] * (amount + 1)           # dp[a] = min coins to make amount a (start as INF)
    dp[0] = 0                           # 0 coins needed to make amount 0

    for a in range(1, amount + 1):      # compute answer for every amount 1..target
        for coin in coins:              # try each coin
            if a - coin >= 0:           # coin can be used only if it doesn't go negative
                dp[a] = min(dp[a], 1 + dp[a - coin])  
                # 1 + dp[a-coin] means:
                # take this coin + best way to make remaining amount (a-coin)

    if dp[amount] == INF:               # if still INF, it means not possible
        return -1
    return dp[amount]                   # otherwise dp[amount] is minimum coins


coins = [1, 2, 5]                       # coin denominations
amount = 5                              # target
print(coin_change_min(coins, amount))   # prints minimum coins needed
```

✅ Fibonacci Sequence (DP) — Explanation

The Fibonacci problem asks us to compute the nth number in the Fibonacci series, where each number is the sum of the previous two (F(n) = F(n−1) + F(n−2)). A simple recursive solution is inefficient because it repeatedly recalculates the same values (overlapping subproblems). Dynamic Programming improves this by storing previously computed results in a table (array). In the bottom-up DP approach, we start from the base cases F(0)=0 and F(1)=1, then iteratively compute each next value using the stored results. This avoids redundant work and reduces the time complexity from exponential to linear O(n), making the computation much faster and efficient.

✅ Longest Common Subsequence (LCS) — Explanation

The Longest Common Subsequence problem finds the length (or actual sequence) of the longest subsequence common to two strings while maintaining order (but not necessarily contiguity). It has optimal substructure and overlapping subproblems, making it suitable for Dynamic Programming. We create a 2D DP table where dp[i][j] represents the LCS length of the first i characters of string X and the first j characters of string Y. If the current characters match, we extend the subsequence by 1; otherwise, we take the maximum from excluding one character from either string. After filling the table, the final answer is in dp[m][n]. This approach runs in O(m×n) time and ensures we systematically compare all prefixes without recomputation.

✅ Coin Change — Total Number of Ways (DP) — Explanation

In the Coin Change (ways) problem, we must count how many different combinations of given coin denominations can form a target amount, assuming unlimited supply of each coin and that order does not matter. Dynamic Programming is used because the number of ways to make a smaller amount helps build the solution for larger amounts. We use a 1D array dp where dp[a] stores the number of ways to make amount a. Starting with dp[0]=1 (one way to make zero), we iterate through each coin and update the array so that each amount accumulates the ways it can be formed using that coin. This builds the solution incrementally and runs in O(n×amount) time.

✅ Coin Change — Minimum Number of Coins (DP) — Explanation

In the Coin Change (minimum coins) problem, the goal is to find the minimum number of coins needed to make a target amount using unlimited coins. This is an optimization problem well suited for Dynamic Programming. We create a DP array where dp[a] represents the minimum coins needed to make amount a. We initialize dp[0]=0 and all others to infinity (meaning not yet reachable). For each amount, we try every coin and update dp[a] = min(dp[a], 1 + dp[a−coin]). This ensures we always keep the best (minimum) solution. If the final value remains infinity, the amount cannot be formed. The algorithm runs in O(n×amount) time and guarantees the optimal minimum.
