```python
class Node:
    def __init__(self, key, freq):
        self.key = key
        self.freq = freq
        self.left = None
        self.right = None
```

```python
# Normal BST insert
def insert(root, key, freq):
    if root is None:
        return Node(key, freq)

    if key < root.key:
        root.left = insert(root.left, key, freq)
    else:
        root.right = insert(root.right, key, freq)

    return root


def greedy_bst_sort(keys, freq):
    # Combine and sort by frequency (descending)
    items = list(zip(keys, freq))
    items.sort(key=lambda x: x[1], reverse=True)

    root = None
    for key, f in items:
        root = insert(root, key, f)

    return root
```

```python
def greedy_bst_max(keys, freq):
    if not keys:
        return None

    # find index of max frequency
    max_index = freq.index(max(freq))

    root = Node(keys[max_index], freq[max_index])

    # split into left and right groups
    left_keys = [k for k in keys if k < root.key]
    left_freq = [freq[keys.index(k)] for k in left_keys]

    right_keys = [k for k in keys if k > root.key]
    right_freq = [freq[keys.index(k)] for k in right_keys]

    # build left first, then right
    root.left = greedy_bst_max(left_keys, left_freq)
    root.right = greedy_bst_max(right_keys, right_freq)

    return root
```

```python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.key, end=" ")
        inorder(root.right)
keys = [10, 20, 5, 8, 15]
freq = [4, 2, 6, 3, 5]

print("Technique 1:")
t1 = greedy_bst_sort(keys, freq)
inorder(t1)

print("\nTechnique 2:")
t2 = greedy_bst_max(keys, freq)
inorder(t2)
```

```python
# ---------- Node ----------
class Node:
    def __init__(self, key, freq):
        self.key = key
        self.freq = freq
        self.left = None
        self.right = None


# ---------- Preorder Traversal ----------
def preorder(root):
    if root is not None:
        print(root.key, end=" ")
        preorder(root.left)
        preorder(root.right)


# ---------- Technique 1: Sort then Insert ----------
def insert(root, key, freq):
    if root is None:
        return Node(key, freq)

    if key < root.key:
        root.left = insert(root.left, key, freq)
    else:
        root.right = insert(root.right, key, freq)

    return root


def greedy_bst_sort(keys, freq):
    items = list(zip(keys, freq))
    items.sort(key=lambda x: x[1], reverse=True)  # sort by freq desc

    root = None
    for key, f in items:
        root = insert(root, key, f)

    return root


# ---------- Technique 2: Max Frequency Root ----------
def greedy_bst_max(keys, freq):
    if not keys:
        return None

    max_index = freq.index(max(freq))
    root = Node(keys[max_index], freq[max_index])

    # left group
    left_keys = [k for k in keys if k < root.key]
    left_freq = [freq[keys.index(k)] for k in left_keys]

    # right group
    right_keys = [k for k in keys if k > root.key]
    right_freq = [freq[keys.index(k)] for k in right_keys]

    root.left = greedy_bst_max(left_keys, left_freq)
    root.right = greedy_bst_max(right_keys, right_freq)

    return root


# ---------- Main ----------
keys = [10, 20, 5, 8, 15]
freq = [4, 2, 6, 3, 5]

# Technique 1
t1 = greedy_bst_sort(keys, freq)
print("Technique 1 Preorder:")
preorder(t1)

print()

# Technique 2
t2 = greedy_bst_max(keys, freq)
print("Technique 2 Preorder:")
preorder(t2)
```
